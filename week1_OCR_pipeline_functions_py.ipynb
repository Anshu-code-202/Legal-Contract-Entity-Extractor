### 3. Implement OCR Pipeline (PDF → Image → OCR → Raw Text)

This section defines functions to convert a PDF into a series of images, apply Tesseract OCR to each image, and then concatenate the extracted text. We will use `pdf2image` for conversion and `pytesseract` for OCR.

```python
import pytesseract
from pdf2image import convert_from_path
import re
import os

# Ensure Tesseract command is correctly set
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

def pdf_to_text(pdf_path):
    """Converts a PDF file to text using OCR via pdf2image and pytesseract."""
    try:
        # Convert PDF pages to a list of PIL images
        pages = convert_from_path(pdf_path)

        full_text = ""
        for i, page in enumerate(pages):
            print(f"Processing page {i+1}/{len(pages)}")
            # Apply Tesseract OCR to each image
            text = pytesseract.image_to_string(page)
            full_text += text + "\n"

        return full_text
    except Exception as e:
        return f"Error during PDF to text conversion: {e}"

print("OCR pipeline function `pdf_to_text` defined.")
```

### 4. Text Cleaning & Noise Removal

After OCR, the raw text often contains noise like extra spaces, newlines, and special characters. This function cleans the text using regular expressions.

```python
def clean_text(text):
    """Cleans the OCR-extracted text by removing extra spaces, newlines, and unwanted characters."""
    # Remove multiple newlines with a single space
    text = re.sub(r'\n+', ' ', text)
    # Remove multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    # Remove characters that are not alphanumeric, basic punctuation, or currency symbols
    # Keeping: letters, numbers, $, ., ,, space
    text = re.sub(
        r'[^a-zA-Z0-9$., ]',
        '',
        text
    )
    # Optional: convert to lowercase if desired for NLP, but often kept original for NER
    # text = text.lower()
    return text.strip()

print("Text cleaning function `clean_text` defined.")
```
